```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("monthly_sales.csv") 
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df["Month"], df["Sales"], marker="o", linewidth=2)
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.title("Monthly Sales Trend")
plt.grid(True)
```
